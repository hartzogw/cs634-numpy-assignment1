```python
import numpy as np
import librosa
from matplotlib import pyplot as plt
from scipy import signal as sign
from scipy.fft import fft, fftshift
```

IMPORT AUDIO FILE

```python
signal, sr = librosa.load("note-to-say-good-bye-221567.mp3")
print(signal, sr)
```

VISUALIZE WAVE DATA

```python
plt.figure(figsize=(15, 5))
#plotting the sampled signal
librosa.display.waveshow(signal)
```

VISUALIZE TIME FREQUENCY DATA

```python

dB = librosa.amplitude_to_db(np.abs(librosa.stft(signal)), ref=np.max)
img = librosa.display.specshow(dB, y_axis='linear', x_axis='time',
                               sr=sr)
```
